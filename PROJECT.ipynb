```python
#1. Display the companies which have cost more than 100   million $.

s.loc[s["Cost (million $)"]>=100,["Company Name","Cost (million $)"]]
```

```python
#2. All successful ISRO missions held under a budget of 50 million dollars.

s.loc[(s["Company Name"]=="ISRO") & (s["Cost (million $)"]<=30) & (s['Mission Status']=="Success"),:]
```

```python
#3.All missions failed
s.loc[s["Mission Status"]=="Failure",:]
```

```python
#4.Cost of all missions failed by ArianeSpace.
Y=s.loc[(s["Company Name"]=="Arianespace" ) & (s["Mission Status"]=="Failure"),"Cost (million $)"]
print(Y)
X=Y.tolist()
AVG=sum(Y)/len(Y)
print("Average cost of failed missions by Arianespace: $",AVG)
```

```python
#5.All the launches succeeded in France
s.loc[(s["Location"].str.contains("France")) & (s["Mission Status"]=="Success"),:]
```

```python
#6.Dataframe in ascending order by cost
s.sort_values(by="Cost (million $)",ascending=False)
```

```python
#7. All the launches in 2010.
s.loc[(s["Date And Time"].str.contains("2010")),:]
```

```python
s.loc[(s.apply(lambda row: row.astype(str).str.contains('France').any(), axis=1)) & (s["Mission Status"]=="Failure"),:]
```

```python
#8. pre launch failures
s.loc[s["Mission Status"]=="Prelaunch Failure",:]
```

```python
#9.All the retired missions launched in 2015
s.loc[(s["Rocket Status"]=="StatusRetired") & (s["Date And Time"].str.contains("2015")),["Rocket Name","Company Name","Date And Time"]]
```

```python
#15. Success rate of launches by ISRO.
Y=s.loc[(s["Company Name"]=="ISRO" ) & (s["Mission Status"]=="Success"),"Cost (million $)"]
Total=s.loc[(s["Company Name"]=="ISRO" ),"Cost (million $)"]
A=len(Y.tolist())
B=len(Total)
Rate=((A/B)*100)
print("Succes rate of ISRO:",round(Rate,2),"%")

#Success rate of launches by SpaceX.
Y1=s.loc[(s["Company Name"]=="SpaceX" ) & (s["Mission Status"]=="Success"),"Cost (million $)"]
Total1=s.loc[(s["Company Name"]=="SpaceX" ),"Cost (million $)"]
A1=len(Y1.tolist())
B1=len(Total1)
Rate1=(A1/B1)*100
print("Succes rate of SpaceX:",round(Rate1,2),"%")
```
